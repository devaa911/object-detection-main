```python
import urllib.request
import torch
import torchvision
from PIL import Image
```

```python
# Download the yolov5s.pt model
url = 'https://github.com/ultralytics/yolov5/releases/download/v6.0/yolov5s.pt'
urllib.request.urlretrieve(url, 'yolov5s.pt')
print("Downloaded fresh yolov5s.pt successfully!")
```

```python
# Load YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True, force_reload=True)
```

```python
# Print Torch and Torchvision versions
print("Torch Version:", torch.__version__)
print("Torchvision Version:", torchvision.__version__)
```

```python
# Load image
img_path = r"object.jpg"  # Replace with your image path
img = Image.open(img_path)

# Run inference
results = model(img)

# Show results
results.show()
```
